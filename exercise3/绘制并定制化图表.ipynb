# 绘制并定制化图表

<p>虽然我们已经用matplotlib绘制了一些图表，但并没有详细介绍他们是怎么工作的，也没有介绍如何对图表设置以及matplotlib其他的许多功能。本章将介绍并练习一些最基本的数据可视化图表类型，如线形图、柱状图、直方图、饼图，以及他们的变形。</p>
<p>Matplotlib是一个强大的工具箱，能满足几乎所有2D和一些3D绘图的需求。通过示例学习matplotlib是其作者推荐的方式。当以后你需要画一个图表时，就可以找到一个相似的例子，然后做些改动来满足新的需求。</p>

## 1、定义图表类型——柱状图、线形图和堆积柱状图

<p>本节将展示基本的图表以及他们的用途。这里介绍的大多数图表都是很常用的，其中有一些是理解数据可视化中更高阶概念的基础。我们从matplotlib.pyplot库的一些常用图表入手，采用一些简单的样本数据开始一些基本的绘图操作，为后面几节内容打基础。</p>

### 代码

```python
from matplotlib.pyplot import *
plot([1,2,3,2,3,2,2,1])
show()
```

<p>这里我们注意到提供给plot()的值是y轴的值。plot()为x轴提供了默认值，在这里为0～7的线性值。</p>
<p>我们来通过为plot()增加一个参数来指定x轴的值。</p>

```python
plot([1,2,3,2,3,2,2,1])
plot([4, 3, 2, 1],[1, 2, 3, 4])
show()
```

<p>从图中可以看到matplotlib通过扩展y轴来适应新的值范围，并且为了让我们能区分出新的图形，自动改变了第二个线条的颜色。</p>
<p>现在让我们基于相同的数据集合多生成一些常见的图表来做一下比较。</p>

```python
from matplotlib.pyplot import *

x=[1,2,3,4]
y=[5,4,3,2]
figure()

subplot(231)
plot(x,y)

subplot(232)
bar(x,y)

subplot(233)
barh(x,y)

subplot(234)
bar(x,y)

y1 = [7, 8, 5, 3]
bar(x, y1, bottom=y, color='r')

subplot(235)
boxplot(x)

subplot(236)
scatter(x,y)

show()
```

### 工作原理

<p>通过调用figure()方法，我们创建出一个新的图表，这个方法也可以添加参数，作为图表的title。如果通过相同的参数（也可以是数字）调用figure()方法，将会激活相应的图表，接下来的绘图操作都在此图表中进行。</p>
<p>接下来，调用subplot(231)方法将图表分割成2*3的网格，也可以用subplot(3，2，1)这种方式调用，参数分别为行、列、图形的标号。</p>
<p>接着用几个简单的命令穿件垂直柱状图（bar()）和水平柱状图（barh()）。对于堆叠柱状图，我们需要把两个柱状图方法调用连在一起。通过设置参数bottom=y，把第二个柱状图和前一个柱状图连接起来形成堆叠柱状图。</p>
<p>通过调用boxplot()方法可以创建箱线图，图中的箱体从下四分位数延伸到上四分位数，并带有一条中值线。</p>
<p>最后创建了一个散点图来让大家对基于点的数据集合有所了解。当一个数据集合中有成千上万的数据点时，散点图很有可能就更合适了。</p>

### 补充说明

<p>现在我们来解释一下箱线图的几个重要特征，默认情况下箱线图会显示以下几部分：</p>
<li>箱体：涵盖四分位数范围的矩形</li>
<li>中值：箱体中间的一条线</li>
<li>箱须：延伸到最大值和最小值的竖直的线，不包括异常值</li>
<li>异常值：箱须之外的点</li>
<p>我们用同一个数据集合来绘制箱线图和直方图，来说明上面的数据项：</p>

```python
from pylab import *

dataset=[113,115,119,121,124,
        124,125,126,126,126,
        127,127,128,129,130,
        130,131,132,133,136]

subplot(121)
boxplot(dataset, vert=False)

subplot(122)
hist(dataset)

show()
```

<p>通过上述对比，我们可以观察到两种图表在数据展现上的差异。箱线图呈现了前面提到的几个统计数据，直方图展示了数据集合在给定范围内的分组情况。</p>

## 2、简单的正弦图和余弦图

<p>本节将复习一下基本的数学函数绘图以及和数学符号相关的内容，比如在标签和绘制的曲线上写上希腊符号</p>

## 代码

<p>首先，我们来计算从-Pi到Pi之间具有相同的线性距离的256个点的正弦值和余弦值，然后把sin(x)值和cos(x)值在同一个图表中绘制出来。</p>

```python
import matplotlib.pyplot as pl
import numpy as np

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y1 = np.cos(x)
y2 = np.sin(x)

pl.plot(x, y1)
pl.plot(x, y2)

pl.show()
```

<p>接下来我们以这个图表为基础，我们可以进一步定制化来增加更多的信息，并且让坐标轴及其边界更精确些。</p>

```python
from pylab import *
import numpy as np

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y1 = np.cos(x)
y2 = np.sin(x)

plot(x, y1)
plot(x, y2)

title("Function $\sin$ and $\cos$")

xlim(-3.0, 3.0)
ylim(-1.0, 1.0)

xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1,0,+1],[r'$-1$', r'$0$', r'$+1$'])

pl.show()
```

<p>上述代码中，我们用如$\sin$或$-\pi$这样的表达式在图表中写上希腊字符。在接下来的章节中会进一步介绍这种Latex语法。</p>

## 3、设置坐标轴长度和范围

<p>本节将演示一些与坐标轴的范围和长度相关的非常有用的信息，这些属性在matplot中都是可以设置的。</p>

### 代码

```python
from matplotlib.pylab import *
# 调用不带参数的axis()方法返回坐标轴的默认值
axis()
# 值分别代表xmin, xmax, ymin, ymax
```

```python
l = [-1, 1, -10, 10]
# 自己设置x轴和y轴的值
axis(l)
```

<p>若如果不使用axis()或者其他参数设置，matplotlib会自动使用最小值，刚好可以让我们在一个图中看到所有的数据点。如果设置axis()，其范围比数据集合中的最大值小，matplotlib会按照设置执行，这样就无法在图中看到所有的数据点了。这可能会因其用户困惑甚至是错误的理解，因为我们认为我们看到了绘制的所有东西。避免这种情况发生的方法之一是调用autoscale()方法，该方法会计算坐标轴的最佳大小以适应数据的显示。</p>
<p>如果想要想相同图形添加新的坐标轴，可以调用matplotlib.pyplot.axes()方法。我们通常会在方法中传入一些属性，例如rect，归一化单位(0,1)下的left、bottom、width、height四个属性，或者axisbg，该参数指定坐标轴的背景颜色。</p>
<p>还有其他一些参数允许我们对新添加的坐标轴进行设置，如sharex/sharey参数，该参数接受其他坐标轴的值并让当前坐标轴共享相同的值；或者polar参数，指定是否使用极坐标轴。</p>
<p>添加新坐标轴在某些情况下非常有用，例如，如果需要用几个不同的视图来表示相同的数据的不同属性值，这就需要在一张图中组合显示多个图表。</p>
<p>如果只想对当前图形增加一条线，可以调用matiplotlib.pyplot.axhline()或者matplotlib.pyplot.axvline()。axhline()和axvline()方法会根据给定的x和y值相应的绘制出相对于坐标轴的水平线和垂直线，这两个方法的参数很相似，axhline()方法比较重要的参数是y向位置、xmin和xmax，axvline()方法比较重要的参数是x向位置、ymin和ymax。</p>

```python
autoscale()
# 默认绘制了一条y=0的直线
axhline()
# 默认绘制了一条x=0的直线
axvline()
# 绘制了一条y=4的直线
axhline(4)
```

<p>类似的，另外两个相关的方法允许我们添加一个跨坐标轴的水平带（矩形），它们是matplotlib.pyplot.axhspan()和matplotlib.pyplot.axvspan()。axhspan()方法必须的ymin和ymax参数指定了水平带的宽度。同理，axvspan()方法必须的xmin和xmax参数指定了垂直带的宽度。</p>

### 补充说明

<p>图形汇总的网格属性默认是关闭的，但打开和定制化很简单。不带参数调用matplotlib.pyplot.grid()会切换网格的显示状态。另外一些控制参数如下：</p>
<li>which:指定绘制的网格刻线类型（major、minor或者both）</li>
<li>axis:指定绘制哪组网格线(both、x或者y)</li>
<p>坐标轴通常由matplotlib.pyplot.axis()控制。坐标轴在内部实现上由几个Python类来表示。其中的一个父类是matplotlib.axes.Axes，包含了操作坐标轴的大多数方法。单一坐标轴由matplotlib.axis.Axis类来表示,matplotlib.axis.XAxis表示x轴，matplotlib.axis.YAxis表示y轴。</p>

<p>下面是几种方法啊可以改变图表中的线条</p>

setp(line, 'linewidth', 1.5)</pre>

<pre>setp(lines, linewidth=1.5)</pre></p>

## 5、设置刻度、刻度标签和网格

<p>本节继续学习如何设置坐标轴和线条属性，并向图形和图表中添加更多的数据。</p>
<p>让我们先来了解以下图形（figure）和子区（subplots）。</p>
<p>在matplotlib中，调用figure()会显式地创建一个图形，表示一个图形用户界面窗口。通过调用plot()或类似的方法会隐式地创建图形。这对于简单的图表没有问题，但是对于更高级的应用，能显式地创建图形并得到示例的引用是非常有用的。</p>
<p>一个图形包括一个或多个子区。子区能以规则网格的方式排列plot()。我们已经使用过subplot()方法，在调用时指定所有plot的行数和列数以及要操作的plot的序号。</p>
<p>如果需要更多的控制，我们需要使用matplotlib.axes.Axes类的坐标轴示例。这样可以把plot防止在图形窗口的任意位置，例如可以把一个小的plot放在一个大的plot中。</p>

### 代码

<p>刻度是图形的一部分，由刻度定位器——指定刻度所在的位置、刻度格式器——指定刻度显示的样式两部分组成。刻度有主刻度和次刻度，默认不显示次刻度。更重要的是，主刻度和次刻度可以被独立地指定位置和格式化。</p>
<p>我们可以使用matplotlib.pyplot.locator_params()方法控制刻度定位器的行为。尽管刻度位置通常会被自行设置，我们仍然可以控制刻度的数目，并且可以在图形比较小的时候使用紧凑视图。</p>

```python
from pylab import *
# 获得现在的坐标轴
ax = gca()
# 设置紧凑视图，并设置刻度间隔
ax.locator_params(tight = True, nbins = 10)

ax.plot(np.random.normal(10, .1, 100))

show()
```

<p>我们可以看到x轴和y轴是如何被切分的，以及数值是如何现实的。我们也可以用locator类完成相同的设置。下面的代码将主定位器设置为10的倍数：<pre>ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(10))</pre></p>
<p>刻度格式器的配置非常简单。格式器规定了值（通常是数字）的显示方式。例如，用matplotlib.ticker.FormatStrFormatter可以方便地指定'%2.1f'或者'%1.1f cm'的格式字符作为刻度标签。</p>
<p>让我们来看一个使用dates模块的例子。matplotlib用浮点值表示日期，其值为从0001-01-01UTC起的天数加1。比如说，0001-01-01 UTC 06:00 的值为1.25。</p>
<p>然后，我们可以用matplotlib.dates.date2num()、matplotlib.dates.num2date()和matplotlib.dates.drange()这样的helper方法对日期进行不同形式的转换。让我们来看一个例子：</p>

## 7、移动轴线到图中央

<p>本节将演示如何移动轴线到图中央。</p>
<p>轴线定义了数据区域的边界，把坐标轴刻度标记连接起来。一共有4个轴线，它们可以放置在任何位置。默认情况下，它们被放置在坐标轴的边界，因此我们会看到数据图表有一个框。</p>

### 代码

<p>为了把轴线移到图中央，需要把其中两个轴线隐藏起来（将color设置为none）。然后，移动另外两个到坐标（0，0）。坐标为数据空间坐标。做法如下：</p>

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-np.pi, np.pi, 500, endpoint=True)
y=np.sin(x)

plt.plot(x, y)

ax=plt.gca()

# 隐藏原坐标轴
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 移动新坐标轴
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# 设置坐标轴刻度位置
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()
```

### 补充说明

<p>这段代码是根据所绘制的图形设计的。我们把轴线移到位置(0, 0)，绘制了一个正弦函数曲线。(0, 0)是图形的中心。这段代码说明了如何把轴线移动到一个特定位置，以及如何去掉不想显示的轴线。</p>
<p>另外，轴线可以被限制在数据结束的地方结束，例如通过调用set_smart_bounds(True)。在这种情况下，matplotlib会尝试以一种复杂的方式设置边界，例如处理颠倒的界限，或者在数据延伸出视图的情况下裁剪线条以适应视图。</p>

## 8、绘制直方图

<p>直方图非常简单，但用它来显示正确的数据也是非常重要的。目前我们仅设计2D直方图。</p>
<p>直方图被用于可视化数据的分布估计。在谈论直方图时我们通常会使用一些术语。表示特定间隔中数据点频率的垂直矩阵称为bin。bin以固定的间隔创建，因此直方图的总面积等于数据点的数量。</p>
<p>直方图可以显示数据的相对频率，而不是使用数据的绝对值。在这种情况下，总面积就等于1。</p>
<p>直方图经常被用在图像处理软件中，一般是作为可视化图像属性（如给定颜色通道上光的分布）的一种方式。这些图像直方图可以进一步应用在计算机视觉算法中来检测峰值，用来辅助边缘检测、图像分割。</p>

### 操作步骤

<p>在绘制直方图时，我们总是想设置正确的bin数量。但是因为没有严格的规则来说明什么时最优bin数量，所以很难做到这一点。在怎样计算bin数量上有几种不同的理论，最简单的一个是基于上取整函数，这是bins(k)等于<pre>ceilling(max(x)-min(x)/h)，</pre>其中x是绘制的数据集合，h为期望的bin宽度。</p>
<p>如果想调用matplotlib.pyplot.hist()来创建直方图，我们需要传入一些参数，下面是一些比较重要的参数：</p>
<li>bins: 可以是一个bin数量的正数值，也可以是表示bin的一个序列。默认是10。</li>
<li>range: bin的范围，当bins参数为序列时，此参数无效。范围外的值将被忽略掉，默认值为None。</li>
<li>normed: 如果为True，直方图的值将进行归一化处理，形成概率密度。默认值为False。</li>
<li>histtype: 默认为bar类型的直方图。其他选项如下：
<ul>barstacked: 用于多种数据的堆叠直方图。</ul><ul>step: 创建未填充的线形图。</ul><ul>stepfilled: 创建默认填充的线形图。histtype的默认值为bar。</ul></li>
<li>align: 用于bin边界之间矩形条的居中设置。默认值为mid，其他值为left和right。</li>
<li>color: 指定直方图的颜色，可以是单一颜色值或者颜色的序列。如果指定了多个数据集合，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色。</li>
<li>orientation: 通过设置orientation为horizontal创建水平直方图。默认值为vertival。</li>
<p>下面代码演示了hist()的用法：</p>

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 10000)

ax = plt.gca()

# 设置bin数量，颜色
ax.hist(x, bins=35, color='B')

ax.set_xlabel('value')
ax.set_ylabel('Frequency')

# title里面是Latex写法
ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' %(mu, sigma))

plt.show()
```

## 9、绘制误差条形图

<p>本节将展示如何创建柱状图以及如何绘制误差条。我们可以用误差条来可视化数据集合汇总的测量不确定度或者指明错误。误差条可以很容易地表示误差偏离数据集合的情况。他们可以显示一个标准差、一个标准误差以及95%的置信区间。因为在表示上没有统一标准，所以我们总是需要显示地表明误差条显示的是哪一种值。实验科学领域的大多数论文都应该在描述数据精度的时候包含误差条。</p>

### 操作步骤

<p>虽然只有两个必选参数——left和height，但是，我们经常会需要使用其他参数，如下：</p>
<li>width: 给定误差条的宽度，默认值是0.8。</li>
<li>bottom: 如果指定了bottom，其值会被加到高度中，默认值为None。</li>
<li>edgecolor: 给定误差条边界颜色。</li>
<li>ecolor: 指定误差条的颜色。</li>
<li>linewidth: 误差条边界宽度，可以设为None（默认值）和0（此时误差条边界将不显示出来）。</li>
<li>orientation: 有vertical和horizontal两个值</li>

<p>一些可选参数（color、edgecolor、linewidth、xerr和yerr）可以是单一值也可以是和误差条数目相同长度的序列。</p>

## 10、绘制饼图

<p>饼图可以描述数值的比例关系，其中每个扇区的弧长大小为其所表示的数量的比例。</p>
<p>饼图很紧凑，看上去很有美感，但是它们也因为难以对数量进行比较而备受批评。饼图的另一个不好的特征是它以特定的角度（视角）的方式和一定的颜色展示数据，这会是我们的感觉有倾向性，从而影响我们对于所呈现数据得出的结论。</p>

### 代码

```python
# 创建一个分裂式饼图
from pylab import *

figure(1, figsize=(6,6))
ax = axes([0.1, 0.1, 0.8, 0.8])

labels = 'Spring', 'Summer', 'Autumn', 'Winter'

x = [15, 30, 45, 10]

explode=(0.1, 0.1, 0.1, 0.1)

pie(x, explode=explode, labels=labels, autopct='%1.1f%%', startangle=67)

title('Rainy days by season')

show()
```

## 11、绘制带填充区域的图表

<p>本节将展示如何对曲线下面的区域或者两个曲线之间的区域进行填充。</p>
<p>借助matplotlib库，我们可以为曲线之间或者曲线下面的区域填充颜色，这样就可以想读者显示那部分区域的值。有些时候，这对读者（观察者）理解给定的特定信息是非常有帮助的。</p>

### 代码

```python
from matplotlib.pyplot import figure,show,gca
import numpy as np

x = np.arange(0.0, 2, 0.01)

y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)

fig = figure()
ax=gca()

ax.plot(x, y1, x, y2, color='black')

ax.fill_between(x, y1, y2, where=y2>y1, facecolor='darkblue', interpolate=True)
ax.fill_between(x, y1, y2, where=y2<y1, facecolor='deeppink', interpolate=True)

ax.set_title('filled between')

show()
```

### 工作原理

<p>在生成预定义间隔的随机信号之后，用常规的plot()方法绘制出这两个信号的图形，然后，调用fill_between()并传入所需的必选参数。</p>
<p>如上图所示，fill_between()方法使用x为定位点，选取y值（y1，y2），然后用几种预定义的颜色绘制出多边形。</p>
<p>用where参数指定一个条件来填充曲线，where参数接收布尔值（可以是表达式），这样就只会填充满足where条件的区域。</p>

### 补充说明

<p>像许多其他用于绘图的函数一样，fill_between()方法也接收许多参数，比如hatch(指定填充的样式替代颜色)和线条选项(linewidth和linestyle)。</p>
<p>另外一个方法是fill_betweenx()，该方法有相似的填充特性，但是它是针对水平曲线的。</p>
<p>更通用的fill()方法提供了对任意多边形填充颜色或者阴影线的功能。</p>

## 12、绘制堆积图

<p>在本节，我们将介绍如何绘制堆积图。在绘制数量时，堆积图表示不同部分在总量中的比重。堆积图不仅能表现总体趋势，也能表现构成总体数量的每一个独立部分的趋势。</p>
<p>我们把全世界的总产能作为我们的总体数量，来绘制不同能源来源对总量的贡献。我们将描绘1973～2014年产能类型的演化。数据包含在ch03-energy-production.csv中。</p>

### 代码

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ch03-energy-production.csv')

columns = ['Coal', 'Natural Gas (Dry)', 'Crude Oil', 'Nuclear Electric Power', 'Biomass Energy', 'Hydroelectric Power','Natural Gas Plant Liquids', 'Wind Energy', 'Geothermal Energy', 'Solar/PV Energy']

colors = ['darkslategray', 'powderblue', 'darkmagenta', 'lightgreen', 'sienna', 'royalblue', 'mistyrose', 'lavender', 'tomato', 'gold']

plt.figure()
polys = plt.stackplot(df['Year'], df[columns].values.T, colors=colors)

rectangles = []

for poly in polys:
    rectangles.append(plt.Rectangle((0,0), 1, 1, fc=poly.get_facecolor()[0]))
legend = plt.legend(rectangles, columns, loc=3)
frame = legend.get_frame()
frame.set_color('white')

plt.title('Primary Energy Production by Source', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Production(Quad BTU)', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(1973, 2014)

plt.show()
```

<p>从上面的图我们可以看到，世界的产能是不断上升的，并且从2005年其进入了一个快速增长极端。我们也能分析出每一个能源类型的演化情况。煤产量缓慢下降，同时天然气和原油开采量持续上升。核能产量也开始在减少。在堆积图的顶部（放大后会更清楚），我们可以看到可再生能源仍然只是全球产能总量中很微不足道的一部分。堆积图在展现这种数据集合方面的表现堪称完美。</p>

### 工作原理

<p>stackplot()方法和plot()方法很相似，只不过它的第二个参数是一个多维数组。该数组的第一维度是填充区域的数目，第二个维度和第一个参数数组相同。在上面这个例子中，df['Year']的形状是(42,)，df[colums].values.T的形状是(10,42)。注意，这里为了确保第二个数组的格式正确，我们使用了转置操作符T。stackplot()返回了一个多边形的列表，我们把它保存在变量polys中。</p>
<p>堆积图目前还不支持图例。因此我们使用plt.Rectangle()方法创建图例矩形。每个矩形的颜色用poly.get_facecolor()[0]指定，这里的ploy是stackplot()方法创建的多边形列表中的元素。</p>
<p>然后，我们使用legend()方法绘制图例，图例矩形作为第一个参数，每种能源类型对应的名字作为第二个参数，第三个参数用来指定图例的位置。设置图例背景为白色。做法通常是首先通过图例对象的get_frame()方法获得图框对象，然后调用其set_color()方法设置颜色。</p>

## 13、绘制带彩色标记的散点图

<p>如果有两个变量，并且想标记出他们之间的相互关系，散点图是解决方案之一。</p>
<p>这种类型的图形也非常有用，它是多维数据可视化进阶的基础，比如绘制散点图矩阵。</p>

### 代码

<p>散点图显示两组数据的值。数据可视化的工作由一组不由线条连接的点完成。每个点的坐标位置由变量的值决定。一个变量是自变量（或成为无关变量），另一个是应变量（或称为相关变量）。应变量通常绘制在y轴上。</p>
<p>下述代码绘制了两幅图，一个是不相关数据，另一个是强正相关数据。</p>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)

y1 = np.random.randn(len(x))

y2 = 1.2 + np.exp(x)

ax1 = plt.subplot(121)
plt.scatter(x, y1, color='indigo', alpha=0.3, edgecolors='white', label='no correl')
plt.xlabel('no correlation')
plt.grid(True)
plt.legend()

ax2 = plt.subplot(122, sharey=ax1, sharex=ax1)
plt.scatter(x, y2, color='green', alpha=0.3, edgecolors='grey', label='correl')
plt.xlabel('strong correlation')
plt.grid(True)
plt.legend()

plt.show()
```

<p>在这里我们也使用了很多参数，如用来设置图形颜色的color、用来设置点状标记（默认是circle）的marker、alpha（alpha透明度）、degecolors（标记的边界颜色）和label（用于图例框）</p>

### 补充说明

<p>散点图通常在应用拟合回归函数之前绘制，用来识别两个变量间的关联。它很好地呈现了相关性的视觉画面，尤其是对非线性关系的数据。matplotlib提供的scatter()函数用来绘制相同长度的一维数组x与y之间的散点图。</p>
